```python
import pandas as pd
from google.colab import files

# Upload the file from your local system
uploaded = files.upload()

# After the file is uploaded, it will be available as a dictionary
# You can get the file name from the uploaded dictionary
annual-enterprise-survey-2023 = list(uploaded.keys())[0]

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(annual-enterprise-2023)

# Now, you can work with the data
n = 9
series = data["annual-enterprise-survey-2023"]
top = series.head(n=n)

# Display the top rows
print(top)
```
