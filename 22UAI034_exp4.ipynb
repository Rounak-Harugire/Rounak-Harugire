```python
import nltk
from nltk.util import ngrams

def generate_ngrams(text, n):
    words = text.split()
    ngram_dict = {}

    for i in range(1, n+1):
        ngram_dict[i] = [" ".join(ngram) for ngram in ngrams(words, i)]

    return ngram_dict

text = "Hi My Name is Kiran and Im Third Year Student Pursuing AIML."

ngram_dict = generate_ngrams(text, 5)

for n, ngrams_list in ngram_dict.items():
    print(f"{n}-grams:")
    for ngram in ngrams_list:
        print(ngram)
    print()
```
