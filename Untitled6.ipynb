```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
import string

# Download necessary resources
nltk.download('punkt')  # Tokenization
nltk.download('stopwords')  # Stopwords
nltk.download('wordnet')  # WordNet
nltk.download('averaged_perceptron_tagger')  # Corrected resource for POS tagging

sample_text = input("Enter text: ")

# Tokenize the text
tokens = word_tokenize(sample_text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Stem the filtered tokens
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

# Define a function to get the wordnet POS tag from a POS tag
def get_wordnet_pos(word):
    from nltk.corpus import wordnet
    tag = nltk.pos_tag([word])[0][1][0].upper()  # Get the first character of the POS tag
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)  # Default to noun if no match

# Lemmatize the filtered tokens based on their POS tags
lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in filtered_tokens]

# Output the results
print("\nOriginal Text:", sample_text)
print("\nTokenized words:", tokens)
print("\nFiltered tokens (stop words removed):", filtered_tokens)
print("\nStemmed words:", stemmed_tokens)
print("\nLemmatized words:", lemmatized_words)
```
