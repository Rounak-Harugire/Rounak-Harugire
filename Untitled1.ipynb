```python
from datetime import datetime
now_date = datetime.now()
print("Current Date & Time is :-",now_date)
```

```python
print("Current Day is :-",now_date.day)
print("Current Month is :-",now_date.month,"October")
print("Current Year is :-",now_date.year)
```

```python
print("Current Hours is :-",now_date.hour,"Hours")
print("Current Minutes is :-",now_date.minute,"Minutes")
print("Current Seconds are :-",now_date.second,"Seconds")
```
